What i will do for cleaning the data ?
# remove nan values 
# get specific columns for processing 
# check if there are duplicated rows 

```python
# remove nan values 
data = data.dropna(axis=1 , how='all')
data.info()
```

```python
data.duplicated().sum()
```

```python
# taking only needed columns 
df = data[['id','subtitle','sq_mt_built','sq_mt_useful','n_rooms','n_bathrooms','n_floors','sq_mt_allotment','rent_price','buy_price','buy_price_by_area','built_year','parking_price']]
df.info()
```

```python
#change column name 
df = df.rename(columns={'subtitle':'city'})
```

```python
df = df[['id','city','sq_mt_built','n_rooms','n_bathrooms','rent_price','buy_price','buy_price_by_area']]
df.info()
```

```python
df.n_rooms.describe()
```

```python
#remove all rows that have no values in the column n_rooms
df.drop(df.index[df.n_rooms==0],axis=0,inplace=True)
```

```python
df.n_rooms.describe()
```

```python
#adding mode into nan cells in n_bathrooms column 

# getting the mode 
df.n_bathrooms.mode()
```

```python
# replacing the nan values by mode 
df.n_bathrooms = df.n_bathrooms.fillna(1)
df.n_bathrooms = df.n_bathrooms.replace(0,1)
```

```python
df.info()
```

```python
 for value in df.rent_price.values:
    if value < 0:
        df.rent_price =0
```

```python
#split city column to (district , city)
df['district'] = df.city.str.split(',',expand=True) [0]
```
